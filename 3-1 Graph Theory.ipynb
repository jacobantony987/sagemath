### 3-1-1 Graph Defintion

Vertices are numbered $0,1,\dots$.

$0 : [1,2,3]$ tells that vertex $0$ is adjacent to vertices $1,2,3$.

```python
G = Graph({0:[1,2,3],1:[0,2,3],2:[0,1],3:[0,1]});G
```

You may skip referring the same edge twice.

```python
H = Graph({0:[1,2,3],1:[2,3]});
G == H
```

### Adding/Deleting Vertices/Edges

```python
G = graphs.CompleteGraph(4);
G.add_vertex(5); G
```

```python
G = graphs.CompleteGraph(4);
G.delete_vertex(1); G
```

```python
G = graphs.CompleteGraph(4);
G.add_edge(1,5); G
```

```python
G = graphs.CompleteGraph(4);
G.delete_edge((1,3)); G
```

### Graph Isomorphism

```python
G1 = Graph({0:[1,2]});
G2 = Graph({1:[0,2]});
G1 == G2
```

```python
G1.is_isomorphic(G2)
```
